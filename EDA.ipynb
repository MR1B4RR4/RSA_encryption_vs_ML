```python
df = pd.read_excel('DF_for_EDA.xlsx')
df.drop('Unnamed: 0', axis=1, inplace=True)
df = df[df['token_length'] % 2 == 0]
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df
```

```python
def join_list(list_):
    new_list = []
    for i in range(0,len(list_),2):
        new_list.append( int(str(list_[i])+str(list_[i+1])) )
    return new_list
```

```python
df['encrypted_eval'] = df['encrypted'].apply(lambda x: eval(x))
df['token_cut_eval'] = df['token_cut'].apply(lambda x: eval(x))
```

```python
df['encrypted_join'] = df['encrypted_eval'].apply(lambda x: join_list(x))
df
```

```python
k = 0
dict_for_model = {}
for i in range(len(df['token_cut_eval'])):
    k += 1
    for j in range(len(df['token_cut_eval'][i])):
        dict_for_model[df['token_cut_eval'][i][j]] = df['encrypted_join'][i][j]
print(k)
```
