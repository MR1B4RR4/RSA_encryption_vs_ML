```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 10)
```

```python
def aplicar_padding(lista, max_len, valor_padding=0):
    return lista + [valor_padding] * (max_len - len(lista))
```

```python
df = pd.read_excel('DF_for_EDA.xlsx')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.dropna(subset='token', inplace=True)
df.drop_duplicates(subset= 'word', inplace=True)
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df['encrypted_eval'] = df['encrypted'].apply(lambda x: eval(x))
df['token_cut_eval'] = df['token_cut'].apply(lambda x: eval(x))
median = np.median([len(i) for i in df['encrypted_eval']])
std = np.std([len(i) for i in df['encrypted_eval']])
df['encrypted_eval_len'] = df['encrypted_eval'].apply(lambda x: len(x))
df = df[df['encrypted_eval_len'] <= median + 2*std]
max_element = max([len(i) for i in df['encrypted_eval']])
df['encrypted_padded'] = df['encrypted_eval'].apply(lambda x: aplicar_padding(x, max_element))
df
```

```python
df = df[['token','encrypted_padded']]
df
```

```python
for i in range(max_element):
    # Asignar el valor correspondiente de la lista a cada nueva columna
    df[f'value_{i+1}'] = df['encrypted_padded'].apply(lambda x: x[i])
```

```python
df = df.drop('encrypted_padded', axis=1)
```

```python
df.to_excel('DF_for_model_training.xlsx')
```

# New EDA for another model Training

```python
df = pd.read_excel('DF_for_EDA.xlsx')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.dropna(subset='token', inplace=True)
df.drop_duplicates(subset= 'word', inplace=True)
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df['encrypted_eval'] = df['encrypted'].apply(lambda x: eval(x))
df['token_cut_eval'] = df['token_cut'].apply(lambda x: eval(x))
median = np.median([len(i) for i in df['encrypted_eval']])
std = np.std([len(i) for i in df['encrypted_eval']])
df['encrypted_eval_len'] = df['encrypted_eval'].apply(lambda x: len(x))
df = df[df['encrypted_eval_len'] <= median + 2*std]
max_element = max([len(i) for i in df['encrypted_eval']])
df['encrypted_padded'] = df['encrypted_eval'].apply(lambda x: aplicar_padding(x, max_element))
```

```python
df = df[['token_cut_eval','encrypted_eval']]
```

```python
# Inicializar listas para almacenar los datos del nuevo DataFrame
tokens = []
encrypteds = []

# Iterar sobre cada fila del DataFrame original
for index, row in df.iterrows():
    # Obtener las listas de tokens y encriptados
    token_list = row['token_cut_eval']
    encrypted_list = row['encrypted_eval']
    
    # Asegurar que las listas tengan el mismo tamaño
    assert len(token_list) == len(encrypted_list), "Las listas no tienen el mismo tamaño"
    
    # Iterar sobre los elementos de las listas y agregarlos a las listas correspondientes
    for token, encrypted in zip(token_list, encrypted_list):
        tokens.append(token)
        encrypteds.append(encrypted)

# Crear el nuevo DataFrame con los datos recopilados
df_resultado = pd.DataFrame({
    "token": tokens,
    "encrypted": encrypteds
})

# Mostrar el nuevo DataFrame para verificar su estructura
df_resultado
```

```python
df_resultado.to_excel('DF_for_model_training_2.xlsx')
```
